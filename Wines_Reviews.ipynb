## EXPLORE DATA (EDA)

```python
# Summarise correlation between features
# With the wine dataset, we can group by country 
# and look at either the summary statistics for all countries' points 
# and price or select the most popular and expensive ones.
```

```python
# Groupby by country
country = df.groupby("country")

# Summary statistic of all countries
country.describe().head()
```

##### This selects the top 5 highest average points among all 44 countries:

```python
country.mean().sort_values(by="points",ascending=False).head(5)
```

```python
plt.figure(figsize=(15,5))
country.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=90)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.show()
```

 Among 44 countries producing wine, US has more than 50,000 types of wine in the wine review dataset, twice as much as the next one in the rank: France - the country famous for its wine. Italy also produces a lot of quality wine, having nearly 20,000 wines open to review.

### Does quantity over quality?

 Let's now take a look at the plot of all 44 countries by its highest rated wine, using the same plotting technique as above:

```python
plt.figure(figsize=(15,5))
country.max().sort_values(by="points",ascending=False)["points"].plot.bar()
plt.xticks(rotation=90)
plt.xlabel("Country of Origin")
plt.ylabel("Highest point of Wines")
plt.show()
```

Australia, US, Portugal, Italy, and France all have 100 points wine. If you notice, Portugal ranks 5th and Australia ranks 9th in the number of wines produces in the dataset, and both countries have less than 8000 types of wine.

##### So now we'll combine all wine reviews into one big text and create a big fat cloud to see which characteristics are most common in these wines.

```python
text = " ".join(review for review in df.description)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

 It seems like black cherry and full-bodied are the most mentioned characteristics, and Cabernet Sauvignon is the most popular of them all. This aligns with the fact that Cabernet Sauvignon "is one of the world's most widely recognized red wine grape varieties. It is grown in nearly every major wine producing country among a diverse spectrum of climates from Canada's Okanagan Valley to Lebanon's Beqaa Valley"

```python
# To explore more lets pour the words into bottle or cup of wine.
# create a shape for wordcloud, first, we need to find a PNG file to become the mask.
```

```python
wine_mask = np.array(Image.open("F:\MLProjects\WinesReviews\images/wine_mask_hd.jpg"))
wine_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
# Transform your mask into a new one that will work with the function:
transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]), np.int32)

for i in range(len(wine_mask)):
    transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
```

```python
# Check the expected result of your mask
transformed_wine_mask
```

```python
# Create a word cloud image
wc = WordCloud(background_color="white", max_words=500,mask= transformed_wine_mask,
               stopwords=stopwords, contour_width=3, contour_color='firebrick')
```

```python
# Generate a wordcloud
wc.generate(text)
```

```python
# store to file
wc.to_file("F:\MLProjects\WinesReviews\images/wine.png")
```

```python
# show
plt.figure(figsize=[15,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### We created a wordcloud in the shape of a wine bottle! It seems like wine descriptions most often mention about black cherry, fruit flavors and full-bodied characteristics of the wine.
